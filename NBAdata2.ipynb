```python
from sportsreference.nba.schedule import Schedule
from datetime import datetime
from sportsreference.nba.boxscore import Boxscores, Boxscore
from sportsreference.nba.teams import Teams
import pandas as pd
import numpy as np
import requests
import json
# from google.colab import drive
# drive.mount('/content/gdrive')

def collect_nba_boxscores(start_year, end_year):
    all_boxscores = {}
    for year in range(start_year, end_year + 1):
        # Get boxscores for each year
        boxscores = Boxscores(datetime(year, 1, 1), datetime(year, 12, 31))
        all_boxscores[year] = boxscores.games

    return all_boxscores
```

```python
# this function takes the above games object and returns a dataframe
## !! takes a long time to run !!
def boxscores_to_df(games_object):
  games = games_object.games
  keys = list(games.keys())
  box_id_list = []
  for key in keys:
    for game in range(len(games[key])):
      box_id_list.append(games[key][game]['boxscore'])
    print(key, box_id_list)
    
  df = pd.DataFrame()
  appended_data = []
  for box_id in box_id_list:
    game_data = Boxscore(box_id).dataframe
    appended_data.append(game_data)
    print(len(appended_data))
  df = pd.concat(appended_data)
  return df
```

```python
nbaCurrent = boxscores_to_df(games)
```
